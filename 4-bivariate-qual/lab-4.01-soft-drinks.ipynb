```python
df = pd.read_csv('https://raw.githubusercontent.com/HoGentTIN/dsai-labs/main/data/Softdrinks.csv', sep=";")
df.head()
```

### Age vs Brand

```python
observed = pd.crosstab(df.Preference, df.Age)

observed.plot(kind='barh', stacked=True)
```

```python
# Chi-squared test for independence based on a contingency table
observed = pd.crosstab(df.Preference, df.Age)
chi2, p, dof, expected = stats.chi2_contingency(observed)

print("Chi-squared       : %.4f" % chi2)
print("Degrees of freedom: %d" % dof)
print("P-value           : %.4f" % p)
```

P Value is groter dan 0.05, dus er is niet voldoende reden om H0 te verwerpen => er bestaat geen verband tussen de leeftijd en de voorkeur voor ons merk of het ander merk.

### Gender vs Brand

```python
observed = pd.crosstab(df.Preference, df.Gender)

observed.plot(kind='barh', stacked=True)
```

```python
# Chi-squared test for independence based on a contingency table
observed = pd.crosstab(df.Preference, df.Gender)
chi2, p, dof, expected = stats.chi2_contingency(observed)

print("Chi-squared       : %.4f" % chi2)
print("Degrees of freedom: %d" % dof)
print("P-value           : %.4f" % p)
```

P value is groter dan 0.05, dus er is niet voldoende reden om H0 te verwerpen => er bestaat geen verband tussen het geslacht en de voorkeur voor ons merk of het ander merk.
